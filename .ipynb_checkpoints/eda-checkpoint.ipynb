```python
import pandas as pd
import numpy as np
```

```python
df_all = pd.read_csv('divorces_2000-2015_translated.csv')
df_all.head()
```

```python
df = pd.read_csv('divorces_2000-2015_translated.csv'
                 ,header=0
#                 ,names=['man','woman','dom'
                 ,usecols=['DOB_partner_man', 'DOB_partner_woman', 'Date_of_marriage']
#                ,parse_dates=True,infer_datetime_format=True
                 )
df.head(1)
```

```python
df_dob = df[['DOB_partner_man','DOB_partner_woman','Date_of_marriage']]
df_dob
```

```python
df_dob.shape
```

```python
df_dob = df_dob.dropna()
df_dob.shape
```

```python
df_dob = df_dob.rename(columns={'DOB_partner_man':'man','DOB_partner_woman':'wom','Date_of_marriage':'dom'})
df_dob.head()
```

```python
df_age = pd.DataFrame({'man'  :pd.to_datetime(df_dob['dom'])
                              -pd.to_datetime(df_dob['man'])
                      ,'wom'  :pd.to_datetime(df_dob['dom'])
                              -pd.to_datetime(df_dob['wom'])
                      ,'dif'  :pd.to_datetime(df_dob['man'])
                              -pd.to_datetime(df_dob['wom'])
                      }
                    )
df_age
```
